```python
# Dependencies
import pandas as pd
```

```python
# load CSV
csv = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
```

```python
# Read with pandas
df = pd.read_csv(csv, low_memory=False)
df.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
new_df = df.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
new_df.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
new_df = new_df.replace({0.0: "No", 1.0: "Yes"})
new_df.head()
```

```python
#Extract row for only those who attended a bootcamp.
attended = new_df.loc[new_df["AttendedBootcamp"]=="Yes", :]
attended.count()
```

```python
bootcamp_name = pd.DataFrame(attended["BootcampName"].value_counts())
bootcamp_name.reset_index(inplace=True)
bootcamp_name.columns = ["BootcampName", "Count"]

#Extract schools with sufficient responses. 
#bootcamp_name = bootcamp_name.loc[(bootcamp_name["Count"] > 9) &
                                  #(bootcamp_name["BootcampName"] != 'Free Code Camp is not a bootcamp - please scroll up and change answer to "no"')
bootcamp_name.head()
```

```python
#Create another DataFrame with two columns: One for the bootcamp name, and one for the number of respondents who recommend it.
attended = attended.replace({"No": 0, "Yes": 1})
bootcamp_recommend = pd.DataFrame(attended.groupby("BootcampName")["BootcampRecommend"].sum())
bootcamp_recommend.reset_index(inplace=True)
bootcamp_recommend.columns = ["BootcampName", "Recommended"]

bootcamp_recommend.head()
```

```python
merged = pd.merge(bootcamp_name, bootcamp_recommend)
merged
```

```python
student_percent = (merged["Recommended"] / merged["Count"]) * 100
student_percent
```

```python
merged["%Recommended"] = (merged["Recommended"] / merged["Count"]) * 100

merged.sort_values(["%Recommended"], ascending=False).round(2)

merged["%Recommended"] = merged["%Recommended"].map("{0:,.2f}%".format)

merged.head()

#student_percent.sort_values(by=[1], ascending=True)
```
